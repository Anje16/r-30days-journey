```r
# 5. REAL-WORLD EXAMPLE: Temperature Conversion
celsius <- 25
fahrenheit <- (celsius * 9/5) + 32

print(paste("Temperature:", celsius, "째C =", fahrenheit, "째F"))

```

```r
# paste() combines multiple text and variable values into a single string
# Syntax of paste(..., sep = " ", collapse = NULL)
# Basic concatenation with default separator (space)

paste("Hello", "World")
# Output: "Hello World"

# Using a custom separator
paste("John", "Doe", sep = "-")
# Output: "John-Doe"

# Combining text and variables
name <- "Alice"
age <- 30
paste("Name:", name, "Age:", age)
# Output: "Name: Alice Age: 30"

# Using collapse with vectors
fruits <- c("apple", "banana", "orange")
paste(fruits, collapse = ", ")
# Output: "apple, banana, orange"

# Custom separator with collapse
paste(fruits, collapse = " | ")
# Output: "apple | banana | orange"
```

```r
# 6. ANOTHER EXAMPLE: Simple Interest Calculator
principal <- 1000
rate <- 0.05  # 5%
time <- 2     # years

interest <- principal * rate * time
total <- principal + interest

print(paste("Principal:", principal))
print(paste("Interest:", interest))
print(paste("Total Amount:", total))
```

```r
# Create variables for your personal info
your_name <- "Anje D"
your_age <- 31
birth_year <- 2026 - your_age

print(paste("My name is", your_name))
print(paste("I was born in", birth_year))
```

```r
# Calculate total shopping cost
item1_price <- 50
item2_price <- 75
item3_price <- 30

total_cost <- item1_price + item2_price + item3_price
discount <- total_cost * 0.10  # 10% discount
final_price <- total_cost - discount

print(paste("Total:", total_cost))
print(paste("Discount:", discount))
print(paste("Final Price:", final_price))
```

```r
# Convert kilometers to miles
kilometers <- 100
miles <- kilometers * 0.621371

print(paste(kilometers, "km =", miles, "miles"))
```
